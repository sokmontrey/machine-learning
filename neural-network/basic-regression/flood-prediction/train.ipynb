```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dir_path = '../../../datasets/kaggle-flood-prediction-comp/'
train_path = dir_path + 'train.csv'
test_path = dir_path + 'test.csv'
submission_path = dir_path + 'submission.csv'
```

```python
df = pd.read_csv(train_path)
df.drop(columns=['id'], inplace=True)
```

```python
plt.figure(figsize=(3, 2))
plt.scatter(df['MonsoonIntensity'], df['FloodProbability'], alpha=0.01, marker='.', color = 'blue')
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, fmt='.1f', annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.plot(corr.iloc[:-1, -1])
plt.ylabel('Correlation with FloodProbability')
plt.xticks(rotation=90)
plt.show()
```
